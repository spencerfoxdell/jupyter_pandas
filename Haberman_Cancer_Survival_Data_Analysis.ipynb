```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataframe = pd.read_csv(r"C:\Users\Spencer_Fox\Desktop\survival_rate_updated2.csv")
dataframe

survived = dataframe.loc[dataframe['Surv_status']==1]
died = dataframe.loc[dataframe['Surv_status']==2]
sns.set_style("whitegrid")

# died
counts, bin_edges = np.histogram(died['axil_nodes'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of nodes - died')
plt.plot(bin_edges[1:], fdf, label='fdf of nodes - died')
plt.legend()

# survived
counts, bin_edges = np.histogram(survived['axil_nodes'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of nodes - survived')
plt.plot(bin_edges[1:], fdf, label='fdf of nodes - survived')
plt.legend()

plt.xlabel("Number of Nodes")
plt.ylabel("Probability")
plt.title("Survived and Died Patients nodes probability")
plt.show()

# Conclusion : 85% of patients which survived / 55% of patients which didn't survived had same amount of nodes.
```

```python
sns.set_style("whitegrid")

# died
counts, bin_edges = np.histogram(died['Age'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of age - died')
plt.plot(bin_edges[1:], fdf, label='fdf of age - died')

# survived
counts, bin_edges = np.histogram(survived['Age'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of age - survived')
plt.plot(bin_edges[1:], fdf, label='fdf of age - survived')

plt.legend()
plt.xlabel("Age")
plt.ylabel("Probability")
plt.title("Survived and Died Patients Age probability")
plt.show()

# Conclusion : There isn't any significant difference is observed between the two classes as far as age feature is taken into account.
```

```python
## op_year
sns.set_style("whitegrid")
grid = sns.FacetGrid(data=dataframe, hue="Surv_status", palette=['r', 'g'], size=4)
grid.map(sns.distplot, "Op_Year")
grid.add_legend()
grid.fig.suptitle("PDFs Operation Year", y = 1.02)
grid.set_ylabels('PDF');
```

```python
## age
sns.set_style("whitegrid")
grid = sns.FacetGrid(data=dataframe, hue="Surv_status", palette=['r', 'g'], size=4).map(sns.distplot, "Age").add_legend()
grid.fig.suptitle("PDFs for Age")
grid.set_ylabels('PDF');
```
