<h1 align="center">Data Description</h1>

The Iris Dataset samples 150 different flowers of the Iris flower species with three separate species (setosa, versicolor, virginica). This dataset is perfectly balanced with 50 flowers for each of the three subgroups. 

<h1 align="center">Attribute Information</h1>

1) SepalLengthCm : The length of the sepal in centimeters

2) SepalWidthCm : The width of the sepal in centimeters

3) PetalLengthCm : The length of the petal in centimeters

4) PetalWidthCm : The width of the petal in centimeters

5) Species : The identification of the species for the flower

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="Purples")
    return styler

dataframe = pd.read_csv(r"C:\Users\Spencer_Fox\Desktop\Iris.csv")
dataframe.head().style.pipe(make_pretty)
```

# Initial Observations
- Iris Dataset has been read in successfully using the read function
- Using the head function, I was able to pull the first 5 rows of the dataset
- Each of the rows appears to be of the "Iris-setosa" species with differing lengths with the exception of the Petal Width (which = .2 cm in width)

```python
# Number of Species in Dataset 
dataframe["Species"].value_counts()
```

```python
def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="Greens")
    return styler

dataframe.describe().style.pipe(make_pretty)
```

- In order to get a better idea of what the dataset looks like, I used the 'describe' function from python pandas to paint the picture
- At first glance, I know the 'Id' column is irrelevant as it is only recording each flower with no other relevant substance
- Across the three species, the largest petal length is 6.9 cm and the smallest petal length is 3.75 cm
- The median SepalWidth of all of the sampled flowers are 3 cm and the median PetalWidth of all the flowers are 1.3 cm
- The variable with the lowest std is the SepalWidth which I can infer that the data is not as spread out in terms of values than that of the other variables

```python
fig = dataframe[dataframe.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='red', label='Setosa')
dataframe[dataframe.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
dataframe[dataframe.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='yellow', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length vs Width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

# Observation of SepalLength vs. SepalWidth Correlation for Species
1. From observing the scatterplot, the Setosa flower is indicating to have the smaller sepal length much more than the other two species
2. All three species appear to be similar in width as the majority of the datapoints lie between 2.5cm and 3.5cm in SepalWidth
3. With the exception of 4 outliers for the Iris-setosa species and 12 outliers for the Iris-virginica species, all of the data points are bound between 4.5cm and 7cm in Sepal Length

```python
fig = dataframe[dataframe.Species=='Iris-setosa'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='red', label='Iris-setosa')
dataframe[dataframe.Species=='Iris-versicolor'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='blue', label='Iris-versicolor',ax=fig)
dataframe[dataframe.Species=='Iris-virginica'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='yellow', label='Iris-virginica', ax=fig)
fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title(" Petal Length vs Width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

# Observation of PetalLength vs. PetalWidth Correlation for Species
1. As opposed to the Sepal correlation comparison, the Petal correlation comparison appears to be a lot more clear with the scatterplot visualization
2. The Iris-setosa species clearly has the smallest petals of the three groups
3. Iris-versicolor petals are not as large as Iris-virginica but are much similar in size compared to the setosa flowers
4. The smallest Iris-versicolor is larger by at least 0.4cm in PetalWidth than the longest Iris-setosa flower

```python
# Length and Width distribution
dataframe = dataframe.drop(["Id"],axis=1)
dataframe.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

# Observation of Histograms of each Variable
1. SepalWidth Histogram is showing a normal distribution figure - the peak (most frequented data value) peaked around the median of the dataset
2. SepalLength Histogram is showing a front heavy distribution - the majority of the data points are of smaller length with few data points making up the larger values
3. PetalLength/PetalWidth don't have a conclusive pattern to the data. Largest peak is found at the smallest bin with a couple other peaks in the middle of the spread.

```python
#pair plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.figure(figsize=(15, 9))
sns.pairplot(data=dataframe, hue='Species')
plt.title('Pairplot of Flower Features')
```

1. By using the pairplot analysis, the outcome of Id vs. Another Variable is as expected. Data points are clustered together in a sequential fashion as each species is grouped together in the dataset.
2. PetalWidth Cm appears to have a strong positive correlation with the other variables in the dataset
3. Both the SepalLength and SepalWidth appear to have no correlation for a relationship with the other variables
4. When analyzing the density plots, the SepalWidth appears to not deviate far from the mean among the three species. All three species peak are very close to one another. Also, the Iris-versicolor and Iris-virginica indicate to have mean value very close to one another.

```python
# box and whisker plot for flower dataset
dataframe = pd.read_csv(r"C:\Users\Spencer_Fox\Desktop\Iris.csv")

plt.boxplot([dataframe.SepalLengthCm, dataframe.SepalWidthCm, dataframe.PetalLengthCm, dataframe.PetalWidthCm],
           labels=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'])
plt.show()
```
