```python
fig = fdf[fdf.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='red', label='Setosa')
fdf[fdf.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
fdf[fdf.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='yellow', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length vs Width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
# Observation of Correlation between SepalLength vs. SepalWidth between each of the species
# From observing the scatterplot, the Setosa flower is indicating to have the smaller sepal length much more than the other two species
# All three species appear to be similar in width as the majority of the datapoints lie between 2.5cm and 3.5cm in SepalWidth
# With the exception of 4 outliers for the Iris-setosa species and 12 outliers for the Iris-virginica species, all of the data points are bound between 4.5cm and 7cm in Sepal Length
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cdf = pd.read_csv(r"C:\Users\Spencer_Fox\Desktop\survival_rate_updated2.csv")
cdf

survived = cdf.loc[cdf['Surv_status']==1]
died = cdf.loc[cdf['Surv_status']==2]
sns.set_style("whitegrid")

# died
counts, bin_edges = np.histogram(died['axil_nodes'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of nodes - died')
plt.plot(bin_edges[1:], fdf, label='fdf of nodes - died')
plt.legend()

# survived
counts, bin_edges = np.histogram(survived['axil_nodes'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of nodes - survived')
plt.plot(bin_edges[1:], fdf, label='fdf of nodes - survived')
plt.legend()

plt.xlabel("Number of Nodes")
plt.ylabel("Probability")
plt.title("Survived and Died Patients nodes probability")
plt.show()

# Conclusion : 85% of patients which survived / 55% of patients which didn't survived had same amount of nodes.
```

```python
sns.set_style("whitegrid")

# died
counts, bin_edges = np.histogram(died['Age'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of age - died')
plt.plot(bin_edges[1:], fdf, label='fdf of age - died')

# survived
counts, bin_edges = np.histogram(survived['Age'], bins=10, density=True)
pdf = counts/(sum(counts))
fdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='pdf of age - survived')
plt.plot(bin_edges[1:], fdf, label='fdf of age - survived')

plt.legend()
plt.xlabel("Age")
plt.ylabel("Probability")
plt.title("Survived and Died Patients Age probability")
plt.show()

# Conclusion : There isn't any significant difference is observed between the two classes as far as age feature is taken into account.
```
